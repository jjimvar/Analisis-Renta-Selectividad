```python
import pandas as pd
#Disgregamos el codigo zonal
#Cargamos datos zonales
dat=pd.read_excel("C:/Users/Usuario/OneDrive/Escritorio/4to GANE/TFG/Data IES.xlsx", sheet_name="X(2)")
cod=pd.read_excel("C:/Users/Usuario/OneDrive/Escritorio/4to GANE/TFG/Data IES.xlsx", sheet_name="Códigos Zonales(2)")

#Creo una varibale con los codigos para matchearla con los de los institutos
dat["codigo"]=dat["Codigo Zonal"].str[:5]
```

```python
# Lista de institutos únicos
institutos = cod["Nombre"].unique()

# DataFrame para almacenar las medias finales de todos los institutos
med_final = pd.DataFrame()

for instituto in institutos:
    # Filtrar por el instituto actual
    tem1 = cod[cod["Nombre"] == instituto]

    # Obtener todos los códigos zonales del instituto (podría haber más de uno)
    codigos_zonales = tem1["Codigo"].astype(str).unique()  # Obtener los códigos únicos
    
    # Filtrar 'dat' por los códigos zonales del instituto
    t2 = dat.copy()
    t2["codigo"] = t2["codigo"].astype(str)
    tem2 = t2[t2["codigo"].isin(codigos_zonales)]  # Filtrar por todos los códigos zonales del instituto
    
    # Variables para las que queremos calcular la media
    var = ['Renta neta media por persona',
           'Renta neta media por hogar', 'Índice de Gini',
           'Edad media de la población',
           'Porcentaje de población menor de 18 años',
           'Porcentaje de población de 65 y más años', 'Tamaño medio del hogar']
    
    # Convertir las columnas a numéricas (por si hay valores no numéricos)
    tem2[var] = tem2[var].apply(pd.to_numeric, errors='coerce')
    
    # Crear un diccionario con las medias de las variables
    m = {x: tem2[x].mean() for x in var}
    
    # Convertir el diccionario en un DataFrame con el nombre del instituto como índice
    med = pd.DataFrame([m], index=[instituto])
    
    # Concatenar las medias del instituto actual al DataFrame final
    med_final = pd.concat([med_final, med])

# Guardar el DataFrame final con las medias para cada instituto
med_final.to_excel("C:/Users/Usuario/OneDrive/Escritorio/4to GANE/TFG/Medias IES (2).xlsx", sheet_name="Medias", index=True)
```
