```python
import pandas as pd
#Disgregamos el codigo zonal
#Cargamos datos zonales
dat=pd.read_excel("C:/Users/Usuario/OneDrive/Escritorio/4to GANE/TFG/Data IES.xlsx", sheet_name="X(1)")
cod=pd.read_excel("C:/Users/Usuario/OneDrive/Escritorio/4to GANE/TFG/Data IES.xlsx", sheet_name="CÃ³digos Zonales(1)")

#Creo una varibale con los codigos para matchearla con los de los institutos
dat["codigo"]=dat["Codigo Zonal"].str[:10]
```

```python
# Lista de institutos Ãºnicos
institutos = cod["Nombre"].unique()

# DataFrame para almacenar las medias finales de todos los institutos
med_final = pd.DataFrame()

for instituto in institutos:
    # Filtrar por el instituto actual
    t1 = cod.copy()
    tem1 = t1[t1["Nombre"] == instituto]

    # Convertir la columna "Codigo" a tipo string si no lo es ya
    tem1["Codigo"] = tem1["Codigo"].astype(str)

    # Separar los cÃ³digos en una lista
    zonas = tem1["Codigo"].str.split(" ").explode()
    
    # Filtrar 'dat' por los cÃ³digos zonales de 'zonas'
    t2 = dat.copy()
    tem2 = t2[t2["codigo"].isin(zonas)]
    
    # Variables para las que queremos calcular la media
    var = ['Renta neta media por persona',
           'Renta neta media por hogar', 'Ãndice de Gini',
           'Edad media de la poblaciÃ³n',
           'Porcentaje de poblaciÃ³n menor de 18 aÃ±os',
           'Porcentaje de poblaciÃ³n de 65 y mÃ¡s aÃ±os', 'TamaÃ±o medio del hogar']
    
    # Convertir las columnas a numÃ©ricas (por si hay valores no numÃ©ricos)
    tem2[var] = tem2[var].apply(pd.to_numeric, errors='coerce')
    
    # Crear un diccionario con las medias de las variables
    m = {x: tem2[x].mean() for x in var}
    
    # Convertir el diccionario en un DataFrame con el nombre del instituto como Ã­ndice
    med = pd.DataFrame([m], index=[instituto])
    
    # Concatenar las medias del instituto actual al DataFrame final
    med_final = pd.concat([med_final, med])

# Mostrar el DataFrame final con las medias para cada instituto
med_final.to_excel("C:/Users/Usuario/OneDrive/Escritorio/4to GANE/TFG/Medias IES.xlsx",sheet_name="Medias(1)", index="True")
```
